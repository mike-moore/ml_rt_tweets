## Import the Needed Modules

```
import os, sys
import pandas as pd
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3' 
sys.path.append(os.path.join(os.getcwd(),'src'))
from TwitterScraper import TwitterScraper, start_stream
from TwitterReader import read_tweets_from_file
```

# Pull Real-Time Data from Twitter

```
output_file = 'my_tweets.csv'
myScraper = TwitterScraper(output_csv_file=output_file, time_limit_seconds=300)
start_stream(myScraper, languages=["en"], track_topics=["Tucker Carlson"])
```

# Display the Most Negative Tweets

```
pd.set_option('display.max_colwidth', None)
df = read_tweets_from_file('my_tweets.csv')
df_neg = df.sort_values(by=['sentiment_lstm'])
df_neg.head(10)
```

```
most_negative_tweets = df_neg['text'].values
print(most_negative_tweets[0:25])
```

# Display the Most Positive Tweets

```
df_pos = df.sort_values(by=['sentiment_lstm'], ascending=False)
df_pos.head(10)
```

```
most_positive_tweets = df_pos['text'].values
print(most_positive_tweets[0:10])
```
